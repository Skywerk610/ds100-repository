```python
import numpy as np
import pandas as pd
data = pd.read_csv('tokyo_2020_swim.csv')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
final_data = data.loc[data['heat'].str.contains('final')]
USAfinal_data = final_data.loc[final_data['team'].str.contains('USA')]
USAfinal_data
```

```python
USAfinal_data['place'] = USAfinal_data['place'].astype(str)
USAmedal_data = USAfinal_data.loc[USAfinal_data['place'].str.contains('1|2|3')]
USAmedal_data
```

```python
solo_events = "name"
USAmedal_solo_data = USAmedal_data.dropna(subset=[solo_events])
USAmedal_solo_data
```

```python
df = USAmedal_solo_data
ave_split = df.filter(like='split_', axis=1).mean(axis=1).round(2)
ave_split
```

```python
USAmedal_solo_data_copy = USAmedal_solo_data.copy()
USAmedal_solo_data_copy['ave_split'] = ave_split
USA_solomedal_data = USAmedal_solo_data_copy[['event','name','time','reaction_time','ave_split','place']]
USA_solomedal_data
```
